```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import clear_output
import time

def create_icosahedron():
    phi = (1 + np.sqrt(5)) / 2  # Golden ratio

    vertices = np.array([
        [-1, phi, 0], [1, phi, 0], [-1, -phi, 0], [1, -phi, 0],
        [0, -1, phi], [0, 1, phi], [0, -1, -phi], [0, 1, -phi],
        [phi, 0, -1], [phi, 0, 1], [-phi, 0, -1], [-phi, 0, 1]
    ])

    # normalise
    vertices /= np.linalg.norm(vertices[0])

    # define the 20 triangular faces
    faces = np.array([
        [0, 11, 5], [0, 5, 1], [0, 1, 7], [0, 7, 10], [0, 10, 11],
        [1, 5, 9], [5, 11, 4], [11, 10, 2], [10, 7, 6], [7, 1, 8],
        [3, 9, 4], [3, 4, 2], [3, 2, 6], [3, 6, 8], [3, 8, 9],
        [4, 9, 5], [2, 4, 11], [6, 2, 10], [8, 6, 7], [9, 8, 1]
    ])

    return vertices, faces

def rotation_matrix(angle, axis):
    axis = axis / np.linalg.norm(axis)
    cos, sin = np.cos(angle), np.sin(angle)
    
    return np.array([
        [cos + axis[0]**2*(1-cos), axis[0]*axis[1]*(1-cos) - axis[2]*sin, axis[0]*axis[2]*(1-cos) + axis[1]*sin],
        [axis[1]*axis[0]*(1-cos) + axis[2]*sin, cos + axis[1]**2*(1-cos), axis[1]*axis[2]*(1-cos) - axis[0]*sin],
        [axis[2]*axis[0]*(1-cos) - axis[1]*sin, axis[2]*axis[1]*(1-cos) + axis[0]*sin, cos + axis[2]**2*(1-cos)]
    ])

# Create figure
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Get icosahedron data
vertices, faces = create_icosahedron()

# Animation loop
for angle in np.arange(0, 4*np.pi, 0.1):
    ax.clear()
    
    # Rotate vertices
    R = rotation_matrix(angle, [1, 1, 1])
    rotated_vertices = vertices @ R.T
    
    # Plot edges
    for face in faces:
        # Get vertices of this face (triangular)
        v0, v1, v2 = rotated_vertices[face]
        
        # Plot triangle edges
        ax.plot3D(*zip(v0, v1), 'b-', linewidth=1, alpha=0.8)
        ax.plot3D(*zip(v1, v2), 'b-', linewidth=1, alpha=0.8)
        ax.plot3D(*zip(v2, v0), 'b-', linewidth=1, alpha=0.8)
    
    # Plot vertices
    ax.scatter(*rotated_vertices.T, c='red', s=50)
    
    # Set equal aspect ratio
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_zlim([-1.5, 1.5])
    
    # Hide axes
    ax.set_axis_off()
    
    # Update display
    clear_output(wait=True)
    display(fig)
    time.sleep(0.01)

plt.close()
```
